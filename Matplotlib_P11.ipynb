```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

#### Line plot

```
x = np.arange(0.0, 10, 0.1)
y = np.sin(x)

fig, ax = plt.subplots(3)
fig.suptitle('Simple line plot')
ax[0].plot(x, y, color = 'red', linestyle='dotted', label='sin(x)')
ax[1].plot(x, np.cos(x), color = 'blue', linestyle='dashed', label='cos(x)')
ax[2].plot(x, np.tan(x), color = 'green', linestyle='solid', label='tan(x)')
plt.legend();

fig.savefig("line_plot.png")
```

#### Scatter plot

```
x = np.arange(0.0, 10, 1)
y = np.sin(x)

fig, ax = plt.subplots(3)
fig.suptitle('Simple line plot')
ax[0].plot(x, y, 'o',color = 'red',  label='sin(x)')
ax[1].plot(x, np.cos(x), 'o', color = 'blue',  label='cos(x)')
ax[2].plot(x, np.tan(x), 'o', color = 'green', label='tan(x)')
plt.legend();

fig.savefig("line_plot.png")
```

#### Bar plot

```
tech_skills = {'Java':50, 'Python':70, 'R': 30, 'SQL': 90} 
skills = list(tech_skills.keys()) 
values = list(tech_skills.values()) 
plt.bar(skills, values, color ='red', width = 0.6) 

plt.xlabel("Programming lang.") 
plt.title("Hutaf's skills") 
plt.show()
```
